Die zu $E$ gehörenden Pfade sind $TTN, TNT$ und $NTT$

$P(E) = p\cdot p \cdot (1-p) + p\cdot (1-p) \cdot p + (1-p)\cdot p \cdot p = 3 \cdot p^2 \cdot(1-p)$

### b)

```python
f = 3*p^2*(1-p); f, expand(f.diff(p))   # die Funktion und die 1. Ableitung
```

```python
löse(f.diff(p))
```

```python
# oder

löse(f.diff(p), set=ja)
```

```python
f2 = f.diff(p, p); f2    # 2. Ableitung
```

```python
f2.subs(p, 0)    # Minimum
```

```python
f2.subs(p, 2/3)   # Maximum
```

```python
f.subs(p, Rational(2, 3))
```

### Zu 4.

### a)
Es handelt sich um die Wahrscheinlichkeit für einen Fehler 1. Art - die Wahrschein-<br>
lichkeit dafür, dass bei wahrer Nullhypothese ein Wert aus dem Ablehnungsbereich <br>
erhalten wird<br><br>
Es ist $H_0: p=0.2$<br><br>
Ablehnungsbereich: $X < 12 \cup X > 28$

```python
vv = BV(100, 0.2)

p = vv.P( X < 12 ) + vv.P( X > 28 )
N(p)
```

### b)
Die Situation wird nur erkannt, wenn sich ein Wert aus dem Ablehnungsbereich <br>
von $H_0$ ergibt, es ist also die entsprechende Wahrscheinlichkeit gefragt

```python
v1 = BV(100, 0.15)

N( v1.P( X < 12 ) + v1.P( X > 28 ) )
```
